```python
import pandas as pd
```

```python
df = pd.read_csv('iris.data', header=None, names=["SL", "SW", "PL", "PW", "type"], index_col=False)
```

```python
def classifyPrediction(row):
    if row["PL"] >= 1 and row["PL"] <= 1.9 and row["PW"] >= 0.1 and row["PW"] <= 0.6:
        return "Iris-setosa"
    elif row["PL"] >= 3 and row["PL"] <= 5.1 and row["PW"] >= 1 and row["PW"] <= 1.8:
        return "Iris-versicolor"
    elif row["PL"] >= 4.5 and row["PL"] <= 6.9 and row["PW"] >= 1.4 and row["PW"] <= 2.5:
        return "Iris-virginica"
```

```python
df["predicted-type"] = df.apply(classifyPrediction, axis=1)
df
```

```python
y_true = df["type"]
x_predicted = df["predicted-type"]

df_confusion = pd.crosstab(y_true, x_predicted, margins=True)
df_confusion
```
