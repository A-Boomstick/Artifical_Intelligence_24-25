```python
import pandas as pd
```

```python
df = pd.read_csv('iris.data', header=None, names=["SL", "SW", "PL", "PW", "type"], index_col=False)
```

```python
def classifyPrediction(row): 
    sWeight = 0
    veWeight = 0
    viWeight = 0

    if row["SL"] >= 4.3 and row["SL"] <= 5.8:
        sWeight += 1
    elif row["SL"] >= 4.9 and row["SL"] <= 7:
        veWeight += 1
    elif row["SL"] >= 4.9 and row["SL"] <= 7.7:
        viWeight += 1

    if row["SW"] >= 2.3 and row["SW"] <= 4.4:
        sWeight += 1
    elif row["SW"] >= 2.0 and row["SW"] <= 3.4:
        veWeight += 1
    elif row["SW"] >= 2.2 and row["SW"] <= 3.8:
        viWeight += 1

    if row["PL"] >= 1 and row["PL"] <= 1.9:
        sWeight += 1
    elif row["PL"] >= 3 and row["PL"] <= 5.1:
        veWeight += 1
    elif row["PL"] >= 4.5 and row["PL"] <= 6.9:
        viWeight += 1

    if row["PW"] >= 0.1 and row["PW"] <= 0.6:
        sWeight += 1
    elif row["PW"] >= 1 and row["PW"] <= 1.8:
        veWeight += 1
    elif row["PW"] >= 1.4 and row["PW"] <= 2.5:
        viWeight += 1

    print(sWeight, " sWeight")
    print(veWeight, " veWeight")
    print(viWeight, " viWeight")

    if sWeight > veWeight and sWeight > viWeight:
        return "Iris-setosa"
    elif veWeight > sWeight and veWeight > viWeight:
        return "Iris-versicolor"
    elif viWeight > sWeight and viWeight > veWeight:
        return "Iris-virginica"
    else:
        return "Iris-setosa"
    
```

```python
df["predicted-type"] = df.apply(classifyPrediction, axis=1)
df
```

```python
y_true = df["type"]
x_predicted = df["predicted-type"]

df_confusion = pd.crosstab(y_true, x_predicted, margins=True)
df_confusion
```
